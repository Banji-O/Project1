### Using Log to modify the scale of values with huge differences will make machine learning model perform better

```python
df_f = pd.DataFrame(
    {'income': [5000, 10000, 15000, 20000, 250000, 30000, 35000, 40000, 45000, 50000, 55000, 60000, 65000, 70000, 75000, 80000, 85000],

     'count': [4371, 3295, 5825, 6047, 6097, 5738, 61000, 5720, 5098, 4991, 5152, 4194, 4411, 3709, 3811, 3766, 3565]
    }
                 )
```

```python
import seaborn as sb
import numpy as np
```

```python
df_f.head()
```

```python
sb.set(rc={'figure.figsize':(11.7, 8.27)})
g = sb.barplot(x = 'income', y = 'count', data = df_f)
g.set_xticklabels(g.get_xticklabels(), rotation = 45,
                  horizontalalignment = 'right');
```

```python
# log normal distribution chart
sb.set(rc={'figure.figsize':(11.7, 8.27)})
g = sb.barplot(x = 'income', y = 'count', data = df_f)
g.set_xticklabels(g.get_xticklabels(), rotation = 45,
                  horizontalalignment = 'right')
g.set(xscale='log');
```
