```python
import pandas as pd
import json


# Function to extract names from a JSON string column
def extract_names(json_str):
    # Load the JSON string into a list of dictionaries
    data_list = json.loads(json_str)
    # Extract the 'name' field from each dictionary and join them into a single string separated by commas
    return ', '.join([item['name'] for item in data_list])

# List of columns to process
columns_to_process = ['genres', 'keywords', 'production_countries', 'production_companies','spoken_languages']  # columns to process

# Process each column in the list
for column in columns_to_process:
    new_column_name = column ## f"{column}_names"
    dff[new_column_name] = dff[column].apply(extract_names)
```

## Modified Z-Score Calculation

```python
# function to find MAD - Mean Absolute Deviation

def get_mad(x):  # x is a data point
    median = np.median(x)
    diff = abs(x - median)  # abs means absolute. Is to conert negative values to positive
    MAD = np.median(diff)
    return MAD
```

```python
MAD = get_mad(dff.revenue)
median = np.median(dff.revenue)
MAD, median
```

```python
# function to find modified z score

def get_modified_z_score(x, median, MAD):
    return 0.6745 * (x - median) / MAD  
```

```python
dff['mod_z_score'] = dff.revenue.apply(lambda x: get_modified_z_score(x, median, MAD))
```

```python
dff.head(2)
```

```python
dff[(dff.mod_z_score > 4) | (dff.mod_z_score < -4)]  # the standard band is  < -3.5 to > 3.5
```

```python
# To select data that are not outliers
mod_z_score_no_outliers = dff[(dff.mod_z_score < 4) & (dff.mod_z_score > -4)] 
```
